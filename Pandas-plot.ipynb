# Pandas plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Series

```python
data = pd.Series(np.random.randn(1000),index=np.arange(1000))
data = data.cumsum()
data.plot()
plt.show()
```

DataFrame

```python
data = pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=list("ABCD"))
data = data.cumsum()
print(data.head()) #head默认会打印前五个数据 
```

```python
data.plot()
plt.show()
```

## scatter 

```python
ax = data.plot.scatter(x='A',y='B',color='DarkBlue',label='Class 1')
data.plot.scatter(x='A',y='C',color='DarkGreen',label='Class 2',ax=ax) #在一张图上打印两张图的数据
plt.show()
```
